```python
# Code to capture face features using pickle file as data storage


import cv2
import dlib
import pickle

# Paths to the model files 
shape_predictor_path = 'shape_predictor_68_face_landmarks.dat'
face_recognition_model_path = 'dlib'

# Initializing dlib's face detector, shape predictor, and face recognition model
face_detector = dlib.get_frontal_face_detector()
shape_predictor = dlib.shape_predictor(shape_predictor_path)
face_recognizer = dlib.face_recognition_model_v1(face_recognition_model_path)
# to Load face database from file
try:
    with open('face_database.pkl', 'rb') as f:
        face_database = pickle.load(f)
except FileNotFoundError:
    face_database = {}

# Function to extract face embeddings
def get_face_embedding(image, face_location):
    shape = shape_predictor(image, face_location)
    return face_recognizer.compute_face_descriptor(image, shape)

# Function to enroll a new face
def enroll_face(name):
    cap = cv2.VideoCapture(0)  # Open the webcam

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Detect faces in the frame
        faces = face_detector(frame)

        if len(faces) > 0:
            face = faces[0]
            face_embedding = get_face_embedding(frame, face)

            # Save the face embedding and name to the database (in this case, a file)
            face_database[name] = [face_embedding]

            # Save the face database to a file
            with open('face_database.pkl', 'wb') as f:
                pickle.dump(face_database, f)

            print(f"Successfully enrolled {name}!")
            break

        # Show the frame
        cv2.imshow("Enroll Face", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Calling functions 
name = input("Enter the name of the person to enroll: ")
enroll_face(name)
```

```python
#Code to recognize face comparing the data stored before

import numpy as np

# Function to recognize a face
def recognize_face():
    cap = cv2.VideoCapture(0)  # Opens the webcam

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Detect faces in the frame
        faces = face_detector(frame)

        if len(faces) > 0:
            for face in faces:
                # Get the embedding for the detected face
                face_embedding = get_face_embedding(frame, face)

                # Compare the face embedding with the database
                recognized_name = "Unknown"
                min_distance = float("inf")

                for name, embeddings in face_database.items():
                    # Compute the distance between the current face and the enrolled faces
                    distance = np.linalg.norm(np.array(embeddings[0]) - np.array(face_embedding))
                    
                    if distance < min_distance:
                        min_distance = distance
                        recognized_name = name

                # Threshold to determine whether it's a match ( this value is tunable )
                if min_distance < 0.6:  # Lower means stricter matching, adjustable as needed
                    print(f"Recognized {recognized_name} with distance: {min_distance}")
                else:
                    print(f"Face not recognized, closest match is {recognized_name} with distance: {min_distance}")

        # Show the frame
        cv2.imshow("Recognize Face", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Run the face recognition function
recognize_face()
```

```python
# to compare embeddings and recognise 

import mysql.connector
import pickle
import numpy as np
import dlib
import cv2
from scipy.spatial.distance import euclidean

# Initialize dlib's face detector and shape predictor
face_detector = dlib.get_frontal_face_detector()
shape_predictor_path = 'shape_predictor_68_face_landmarks.dat'
face_recognition_model_path = 'dlib'  # Ensure correct model path
shape_predictor = dlib.shape_predictor(shape_predictor_path)
face_recognizer = dlib.face_recognition_model_v1(face_recognition_model_path)

# Connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Naaz@123",
    database="school_database"
)
cursor = conn.cursor()

# Function to align face and get embedding
def get_face_embedding(image, face_location):
    shape = shape_predictor(image, face_location)
    aligned_face = dlib.get_face_chip(image, shape)  # Align the face
    return face_recognizer.compute_face_descriptor(aligned_face)

def get_embeddings_from_db():
    cursor.execute("SELECT student_id, student_name, student_embeddings FROM students")
    results = cursor.fetchall()

    embeddings_dict = {}
    for student_id, student_name, embeddings_blob in results:
        student_embeddings = pickle.loads(embeddings_blob)
        embeddings_dict[student_id] = {
            'name': student_name,
            'embedding': np.array(student_embeddings)
        }
    return embeddings_dict

# Function to capture and extract embedding from webcam
def capture_face():
    cap = cv2.VideoCapture(0)  # Opens the webcam

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        faces = face_detector(frame)

        if len(faces) > 0:
            face = faces[0]
            face_embedding = get_face_embedding(frame, face)

            cap.release()
            cv2.destroyAllWindows()
            return np.array(face_embedding)

        cv2.imshow("Capture Face", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Function to recognize face using Euclidean distance
def recognize_face(new_embedding, stored_embeddings):
    min_distance = float('inf')
    recognized_name = "Unknown"  # Default for unknown faces
    threshold = 0.6  # Euclidean distance threshold for recognition

    for student_id, data in stored_embeddings.items():
        distance = euclidean(new_embedding, data['embedding'])
        if distance < min_distance:
            min_distance = distance
            recognized_name = data['name']

    # If the minimum distance is still higher than the threshold, return "Unknown"
    if min_distance < threshold:
        return recognized_name
    else:
        return "Unknown"

# Main script
stored_embeddings = get_embeddings_from_db()
new_face_embedding = capture_face()

if new_face_embedding is not None:
    recognized_name = recognize_face(new_face_embedding, stored_embeddings)
    print(f"Recognized face as: {recognized_name}")
else:
    print("No face captured.")

# Close the MySQL connection
conn.close()
```
